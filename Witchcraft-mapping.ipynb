```python
import pandas as pd
import random

```

```python
#loading in the accused CSV
df_accused = pd.read_csv('Witchcraft-csvs/WDB_Accused.csv')
```

```python
#checking out the shape. 3219 rows, if I remove the ones with no county, there sould be 3219-115 = 3104
df_accused.shape
```

```python
#remove all the empty counties 
df_accused = df_accused.dropna(subset=['Res_county'])
```

```python
#checking the shape,there are indeed 3104 counties. Meaning I dropped only the rows with na counties
df_accused.shape
```

```python
#checking all the counties to make sure that I have them all
len(df_accused['Res_county'].unique())
#Yep, I've got 34
```

```python
#now to load in the dataset with the lat and long 
df_lat_long = pd.read_csv('Witchcraft-csvs/Lat_long_witches_csv.csv')
```

```python
#removing those extra columns just in case 
df_lat_long = df_lat_long.drop(columns = ['Unnamed: 4', 'Unnamed: 5'])
```

```python
lat_jitter = .08

# Add +/- lat_jitter to the latitude
def add_jitter_lat(lat):
    return lat + random.random() * lat_jitter * 2 - lat_jitter

# Send every latitude to the add_jitter_lat function
# and put the result in a new column called jittered_lat
map_data['jittered_lat'] = map_data.Latitude.apply(add_jitter_lat)



```

```python
long_jitter = .08

# Add +/- lat_jitter to the latitude
def add_jitter_long(long):
    return long + random.random() * long_jitter * 2 - long_jitter

# Send every latitude to the add_jitter_lat function
# and put the result in a new column called jittered_lat
map_data['jittered_long'] = map_data.Longitude.apply(add_jitter_long)

```
